```python
def square():
    make_turtle()
    for i in range(4):
        forward(50)
        left(90)
square()
```

```python
def square2(length):
    for i in range(4):
        forward(length)
        left(90)

make_turtle()
square2(20)
square2(50)
# adding a parameter to the function is called generalization
```

```python
def polygon(sides, length):
    angle = 360 / sides
    for i in range(sides):
        forward(length)
        left(angle)

make_turtle()
polygon(6, 25)
```

```python
import math

def circle(radius):
    circumference = 2 * math.pi * radius
    n = 30
    length = circumference / n
    polygon(n, length)
```

```python
make_turtle(delay=0.02)
circle(25)
```

```python
def polyline(n, length, angle):
    for i in range(n):
        forward(length)
        left(angle)

def polygon2(n, length):
    angle = 360 / n
    polyline(n, length, angle)

def arc(radius, angle):
    arc_length = 2 * math.pi * radius * angle / 360
    n = 30
    length = arc_length / n
    step_angle = angle / n
    polyline(n, length, step_angle)
```

```python
def circle2(radius):
    arc(radius, 360)
```

```python
make_turtle(delay=0)
polygon2(n=20, length=9)
arc(90,60)
circle2(10)
```
