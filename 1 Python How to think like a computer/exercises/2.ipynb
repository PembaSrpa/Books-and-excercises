### Why is class a keyword?

class is a keyword because it is a reserved identifier used by the interpretor to define a blueprint and structure for creating new objects.

### Why can’t variable names be keywords

Variable names cannot be keywords because keywords have predefined, hardcoded meanings in the Python interpreter.

### Why is it bad to use int, float, and string as variable names?

It is bad practice because these are built-in functions/types in Python; assigning a value to them "shadows" the original names, making it impossible to use the actual int(), float(), or str() functions later in your code.

### What are the built-in functions in Python?

Built-in functions are pre-installed, globally available functions in Python that can be used without importing any external modules.

Common examples include:
- Type Conversion: int(), float(), str(), list(), dict(), set().
- Math & Logic: abs(), min(), max(), sum(), round(), pow().
- Data Processing: len(), sorted(), filter(), map(), zip(), enumerate().
- Input/Output: print(), input(), open().
- Object Inspection: type(), id(), dir(), help(), isinstance()

### What variables and functions are in the math module?

Its variables include well-known constants like math.pi, math.e, and math.tau, as well as floating-point values for infinity and "Not a Number" (NaN). Its functions are categorized into rounding tools like ceil() and floor(), power and logarithmic functions such as sqrt(), pow(), and log(), and a full suite of trigonometric operations like sin(), cos(), and radians().

### Other than math, what modules are considered core Python?

System and file handling modules like os, sys, and shutil. \
Data Type modules like datetime, collections, re. \
For Logic and Math, modules like random, statistics, and itertools. \
And finally Data Interchange modules like json, csv and sqlite3.

## Python interpretor as a calculator

```python
# The volume of a sphere with radius r is (4/3)πr3. What is the volume of a sphere with radius 5? Start with a variable named radius and then assign the result to a variable named volume. Display the result. Add comments to indicate that radius is in centimeters and volume is in cubic centimeters.

import math

radius = 5
volume = (4/3) * math.pi * radius**3

volume
```

```python
# A rule of trigonometry says that for any value of x, cosx2+sinx2=1. Let’s see if it’s true for a specific value of x like 42. Create a variable named x with this value. Then use math.cos and math.sin to compute the sine and cosine of x, and the sum of their squares. The result should be close to 1. It might not be exactly 1 because floating-point arithmetic is not exact—it is only approximately correct.

import math

x = 42
result = math.cos(x)**2 + math.sin(x)**2

result
```

```python
# In addition to pi, the other variable defined in the math module is e, which represents the base of the natural logarithm, written in math notation as e. Now let’s compute e2 three ways:

import math

# Use math.e and the exponentiation operator (**).

res1 = math.e ** 2

# Use math.pow to raise math.e to the power 2.

res2 = math.pow(math.e, 2)

# Use math.exp, which takes as an argument a value, x, and computes ex.

res3 = math.exp(2)

res1, res2, res3
```

### You might notice that the last result is slightly different from the other two. See if you can find out which is correct.
