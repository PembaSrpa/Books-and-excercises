```python
# Write a function called rectangle that draws a rectangle with given side lengths.

from jupyturtle import make_turtle, forward, left, penup, pendown

def jump(length):
    """
    move the turtle forward without drawing
    """
    penup()
    forward(length)
    pendown()

def rectangle(width, height):
    """
    draw a rectangle with given width and height
    """
    for i in range(2):
        forward(width)
        left(90)
        forward(height)
        left(90)

make_turtle()
rectangle(40, 20)
jump(-50)
rectangle(25, 25)
jump(100)
rectangle(25, 25)
```

```python
# Write a function called rhombus that draws a rhombus with a given side length and a given interior angle.

def rhombus(length, angle):
    """
    draw a rhombus with given side length and interior angle
    """
    for i in range(2):
        forward(length)
        left(angle)
        forward(length)
        left(180 - angle)

make_turtle()
rhombus(50, 60)
```

```python
make_turtle()
rhombus(40, 60)
jump(-50)
rectangle(40, 40)
jump(100)
rhombus(60, 60)
jump(-120)
rectangle(-50, 40)
```

```python
# Write an appropriately general set of functions that can draw shapes like this. Hint: write a function called triangle that draws one triangular segment, and then a function called draw_pie that uses triangle.

import math
from jupyturtle import forward, left, make_turtle

def triangle(r, angle):
    """
    draw a triangle with given radius and interior angle
    """
    edge = 2 * r * math.sin(math.radians(angle / 2))
    base_angle = (180 - angle) / 2

    forward(r)
    left(180 - base_angle)
    forward(edge)
    left(180 - base_angle)
    forward(r)
    left(180 - angle)

def draw_pie(n, r):
    """
    draw a pie with given number of segments and radius
    """
    angle = 360 / n
    for i in range(n):
        triangle(r, angle)
        left(angle)

make_turtle()
draw_pie(6, 50)
```

```python
make_turtle()
jump(-100)
draw_pie(5, 50)
jump(100)
draw_pie(6, 50)
jump(100)
draw_pie(7, 50)
```

```python
# Write an appropriately general set of functions that can draw flowers like this.
# Hint: use arc to write a function called petal that draws one flower petal.

import math
from jupyturtle import forward, left, make_turtle

def polyline(n, length, angle):
    """
    Draw n line segments with the given length and angle (in degrees) between them.
    """
    for i in range(n):
        forward(length)
        left(angle)

def arc(r, angle):
    """
    Draw an arc with the given radius and angle (in degrees).
    """
    arc_length = 2 * math.pi * r * angle / 360
    n = int(arc_length / 4) + 1
    length = arc_length / n
    step_angle = angle / n
    polyline(n, length, step_angle)

def petal(r, angle):
    """Draws a petal using two arcs.
    r: radius of the arcs
    angle: angle (degrees) that subtends the arcs
    """
    for i in range(2):
        arc(r, angle)
        left(180 - angle)

def flower(n, r, angle):
    """Draws a flower with n petals.
    n: number of petals
    r: radius of the arcs
    angle: angle of the arcs
    """
    for i in range(n):
        petal(r, angle)
        left(360 / n)

make_turtle()
flower(7, 20, 60)
```
