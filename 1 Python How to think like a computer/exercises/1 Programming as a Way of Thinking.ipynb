### What is the order of operations in Python?
Python follows PEMDAS:

- Parentheses ()
- Exponents **
- Multiplication *, Division /, Flooring //, Modulus % (Left-to-right)
- Addition +, Subtraction - (Left-to-right)

### The round function, which we used to round a floating-point number to the nearest whole number, can take a second argument.“What are the arguments of the round function?”

The round() function takes two arguments: number and ndigits.

```python
# round(number, ndigits)
# To round π to three decimal places:

round(3.14159, 3)
```

### What is the modulus operator in Python?

The modulus operator % returns the remainder of a division.

```python
7 % 2
```

```python
# If a number ends in 0.5, does Python round up or down?

round(0.5)
```

### You can use a minus sign to make a negative number like -2. What happens if you put a plus sign before a number? What about 2++2?

Using a plus sign before a number like +2 is valid but has no effect, while 2++2 evaluates to 4 because Python treats the second plus as a unary operator (2+(+2)).

```python
+2
```

```python
2++2
```

### What happens if you have two values with no operator between them, like 4 2? 
Writing two values without an operator, like 4 2, results in a SyntaxError because Python cannot interpret the relationship between them.

```python
4 2
```

### If you call a function like round(42.5), what happens if you leave out one or both parentheses?

If you omit both parentheses, round just references the function object; if you omit only one, like round(42.5, Python returns a SyntaxError for incomplete syntax.

## Arithmetic expressions:

```python
# How many seconds are there in 42 minutes 42 seconds?
(42*60)+42
```

```python
# How many miles are there in 10 kilometers? Hint: there are 1.61 kilometers in a mile.
10*(1/1.61)
```

```python
# If you run a 10 kilometer race in 42 minutes 42 seconds, what is your average pace in seconds per mile?
((42*60)+42)/(10*(1/1.61))
```

```python
# What is your average pace in minutes and seconds per mile?
pace_in_sec = ((42*60)+42)/(10*(1/1.61))
print("minutes:",pace_in_sec//60)
print("seconds:",pace_in_sec%60)
```

```python
# What is your average speed in miles per hour?
miles = 10/1.61
hrs = (42 + 42 / 60) / 60
mph = miles/hrs
mph
```
