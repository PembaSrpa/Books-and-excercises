### What are the differences between ephemeral and persistent programs?

- Ephemeral programs run for a short time and store data in volatile memory (RAM). When the program ends, all data is lost.

- Persistent programs run for long periods or repeatedly and store data in non-volatile memory (Hard Drive/SSD). The data remains available even after the program or the computer is restarted.

### What are some examples of persistent programs?

Common examples include:

- Operating Systems: They maintain settings and files across reboots.

- Database Management Systems (DBMS): Such as PostgreSQL or SQLite, which store records permanently.

- Web Servers: Which serve files stored on a disk.

- Word Processors: Like Microsoft Word, which save documents to the file system.

### What’s the difference between a relative path and an absolute path?

- Absolute path: Specifies a location starting from the root directory of the file system (e.g., /Users/name/docs/file.txt or C:\Users\name\file.txt).

- Relative path: Specifies a location starting from the current working directory (e.g., docs/file.txt).

Why does the yaml module have functions called load and safe_load?

The yaml.load() function is potentially insecure because it can execute arbitrary Python code embedded in a YAML file. The yaml.safe_load() function restricts the parser to standard YAML tags, preventing the execution of malicious code and making it the "Best-in-Class" choice for loading untrusted data.

### When I write a Python shelf, what are the files with suffixes .dat and .dir?

- .dat file: This file contains the actual serialized data objects stored in the shelf.

- .dir file: This is an index file used by the underlying database management (dbm) library to quickly locate where specific keys and values are stored within the .dat file.

### Other than key-values stores, what other kinds of databases are there?

- Relational Databases (SQL): Store data in tables with predefined schemas (e.g., MySQL).

- Document Databases: Store data as JSON-like documents (e.g., MongoDB).

- Graph Databases: Focus on relationships between data points (e.g., Neo4j).

- Time-Series Databases: Optimized for time-stamped data (e.g., InfluxDB).

### When I read a file, what’s the difference between binary mode and text mode?

- Text mode ('r'): Python automatically handles character encoding (like UTF-8) and translates platform-specific line endings (\n on Unix vs \r\n on Windows) into a single \n.

- Binary mode ('rb'): Python reads the raw bytes exactly as they are stored on the disk without any translation or encoding.

### What are the differences between a bytes object and a string?

- String: A sequence of Unicode characters intended for human-readable text.

- Bytes object: A sequence of 8-bit integers ranging from 0 to 255. It represents machine-readable data or encoded text.

### What is a hash function?

A hash function is an algorithm that takes an input (or "message") of any size and returns a fixed-size string of bytes. The output is deterministic, meaning the same input always produces the same output.
What is an MD5 digest?

An MD5 digest is the 128-bit hash value produced by the MD5 algorithm. It is typically represented as a 32-character hexadecimal string. While no longer considered secure for cryptography, it is still widely used to verify that a file has not been corrupted during transfer.

```python
# Identify Image files

import os

def is_image(path, extensions):
    # splitext returns a tuple like ('/path/to/image', '.jpg')
    _, ext = os.path.splitext(path)
    # Check if the extension (converted to lowercase) is in our list
    return ext.lower() in extensions
```

```python
# Hash and Store Paths

import hashlib

def md5_digest(filename):
    """Computes the MD5 hash of a file's content."""
    with open(filename, 'rb') as f:
        data = f.read()
    return hashlib.md5(data).hexdigest()

def add_path(path, shelf):
    digest = md5_digest(path)
    if digest not in shelf:
        shelf[digest] = [path]
    else:
        # Retrieve, mutate, and reassign
        path_list = shelf[digest]
        path_list.append(path)
        shelf[digest] = path_list
```

```python
# Process and walk

import shelve

# Global or config-based extensions
IMAGE_EXTENSIONS = ['.jpg', '.jpeg', '.png', '.gif', 'txt']

def process_path(path):
    if is_image(path, IMAGE_EXTENSIONS):
        add_path(path, db)

def walk(dirname, visit_func):
    """Recursively walks a directory and applies a function to files."""
    for name in os.listdir(dirname):
        path = os.path.join(dirname, name)
        if os.path.isfile(path):
            visit_func(path)
        elif os.path.isdir(path):
            walk(path, visit_func)
```

```python
def same_contents(path1, path2):
    """Final verification that two files are identical byte-for-byte."""
    with open(path1, 'rb') as f1, open(path2, 'rb') as f2:
        return f1.read() == f2.read()

# Execute the search
with shelve.open('photos/digests', 'n') as db:
    walk('photos', process_path)

    print("Checking for duplicates...")
    for digest, paths in db.items():
        if len(paths) > 1:
            print(f"\nPotential duplicates found for digest {digest}:")
            for p in paths:
                print(f" {p}")

            # Verify the first two in the list
            if same_contents(paths[0], paths[1]):
                print("CONFIRMED: Files contain identical data.")
```
