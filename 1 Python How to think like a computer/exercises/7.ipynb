# Iteration and Search

```python
def uses_any(word, letters):
    """
    take a word and a string of letters, return True if any letter in the word is in the letters string
    """
    for letter in word.lower():
        if letter in letters.lower():
            return True
    return False

def uses_any_incorrect(word, letters):
    """
    if the first letter of the word is not in the letters string, the else block executes immediately and returns False
    """
    for letter in word.lower():
        if letter in letters.lower():
            return True
        else:
            return False
```

```python
def uses_none(word, forbidden):
    """Checks whether a word avoid forbidden letters.
    >>> uses_none('banana', 'xyz')
    True
    >>> uses_none('apple', 'efg')
    False
    """
    for letter in word.lower():
        if letter in forbidden.lower():
            return False
    return True
```

```python
def uses_only(word, available):
    """Checks whether a word uses only the available letters.
    >>> uses_only('banana', 'ban')
    True
    >>> uses_only('apple', 'apl')
    False
    """
    for letter in word.lower():
        if letter not in available.lower():
            return False
    return True
```

```python
def uses_all(word, required):
    """Checks whether a word uses all the required letters.
    >>> uses_all('banana', 'ban')
    True
    >>> uses_all('apple', 'apl')
    False
    """
    for letter in required.lower():
        if letter not in word.lower():
            return False
    return True
```

```python
# Write a function called check_word that checks whether a given word is acceptable. It should take as parameters the word to check, a string of seven available letters, and a string containing the single required letter.

def check_word(word, available, required):
    """Check whether a word is acceptable.
    >>> check_word('color', 'ACDLORT', 'R')
    True
    >>> check_word('ratatat', 'ACDLORT', 'R')
    True
    >>> check_word('rat', 'ACDLORT', 'R')
    False
    >>> check_word('told', 'ACDLORT', 'R')
    False
    >>> check_word('bee', 'ACDLORT', 'R')
    False
    """
    if len(word) < 4:
        return False
    return uses_only(word, available) and uses_all(word, required)
```

```python
# Write a function called score_word that takes a word and a string of available lessons and returns its score. You can assume that the word is acceptable.

def word_score(word, available):
    """Compute the score for an acceptable word.
    >>> word_score('card', 'ACDLORT')
    1
    >>> word_score('color', 'ACDLORT')
    5
    >>> word_score('cartload', 'ACDLORT')
    15
    """
    if len(word) == 4:
        score = 1
    else:
        score = len(word)

    is_pangram = True
    for letter in available.lower():
        if letter not in word.lower():
            is_pangram = False
            break

    if is_pangram:
        score += 7

    return score
```

```python
def uses_none(word, forbidden):
    """Checks whether a word avoids forbidden letters.
    >>> uses_none('banana', 'xyz')
    True
    >>> uses_none('apple', 'efg')
    False
    >>> uses_none('', 'abc')
    True
    """
    return not uses_any(word, forbidden)

def uses_all(word, required):
    """Checks whether a word uses all required letters."""
    return uses_only(required, word)
```
